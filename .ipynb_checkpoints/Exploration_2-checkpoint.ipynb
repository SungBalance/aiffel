# import

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

# Iris

## load dataset

```python
iris = load_iris()
iris_data = iris.data
iris_label = iris.target
```

## preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, 
                                                    iris_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
```

## Model

### Decision Tree

```python
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=24)
random_forest.fit(X_train, y_train)
y_pred_rf = random_forest.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

### Support Vector Machine

```python
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print(classification_report(y_test, y_pred_svm))
```

### Stochastic Gradient Descent Classifier

```python
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier(loss="hinge", penalty="l2")
sgd_model.fit(X_train, y_train)
y_pred_sgd = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred_sgd))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(solver='lbfgs', max_iter=150)
logistic_model.fit(X_train, y_train)
y_pred_log = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

# digits

## load dataset

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(digits.data[0].reshape(8, 8), cmap='gray')
plt.axis('off')
plt.show()
```

```python
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(digits.data[i].reshape(8, 8), cmap='gray')
    plt.axis('off')
plt.show()
```

```python
digits_label = digits.target
print(digits_label.shape)
digits_label[:20]
```

## preprocessing

데이터가 3인지 아닌지를 확인하기 위해 정답 레이블 수정

```python
new_label = [3 if i == 3 else 0 for i in digits_label]
new_label[:20]
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, 
                                                    new_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
```

## building a model

```python
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

데이터의 불균형으로 인해 모든 답을 0으로 지정해도 90% 이상의 정확도를 보임

```python
fake_pred = [0] * len(y_pred)

accuracy = accuracy_score(y_test, fake_pred)
accuracy
```

## Confusion Matrix

- Precision: TP/(FP+TP)
- Sensitivity(Recall): TP/(FN+TP)
- F1 score: 2/(1/Recall + 1/Precision)
- Accuracy = (TP+TN)/(TP+TN+FP+FN)

precision | 음성인데 양성으로 판단하는 경우가 적음: 음성을 놓치면 안 됨(스팸)

recall | 양성인데 음성으로 판단하는 경우가 적음: 양성을 놓치면 안 됨(암 진단)

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix(y_test, fake_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, fake_pred))
```

# project: Digits

### Load dataset

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.target_names
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(digits.data[i].reshape(8, 8), cmap='gray')
    plt.axis('off')
plt.show()
```

### preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
```

### build models

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred_dt = decision_tree.predict(X_test)

# Random Forest
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=24)
random_forest.fit(X_train, y_train)
y_pred_rf = random_forest.predict(X_test)

# Support Vector Machine
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

# Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier(loss="hinge", penalty="l2")
sgd_model.fit(X_train, y_train)
y_pred_sgd = sgd_model.predict(X_test)

# Logistic Regression
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(solver='saga', max_iter=3000)
logistic_model.fit(X_train, y_train)
y_pred_log = sgd_model.predict(X_test)
```

### predict

손글씨 인식은 recall과 precision 모두 고려해야 하므로 f1-score을 기준으로 하여 SVM 모델을 채택할 수 있다.

```python
from sklearn.metrics import classification_report

print("\n----------------------------------------")
print("Decision Tree\n")
print(classification_report(y_test, y_pred_dt))

print("\n----------------------------------------")
print("Random Forest\n")
print(classification_report(y_test, y_pred_rf))

print("\n----------------------------------------")
print("SVM\n")
print(classification_report(y_test, y_pred_svm))

print("\n----------------------------------------")
print("SGD Classifier\n")
print(classification_report(y_test, y_pred_sgd))

print("\n----------------------------------------")
print("Logistic Regression\n")
print(classification_report(y_test, y_pred_log))
```
